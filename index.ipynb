**NOTE: Technical Skills, Independent Learning, and Collaboration** This course focuses on three distinct learning objectives. The first, and most obvious objective, is to develop practical software development technical skills. However, and perhaps even more importantly, this course strives to develop effective learning strategies and collaboration skills. The ultimate goal is not to learn a long list of detailed facts presented in class, but rather to become an effective life-long learner and a dependable team member. For this reason, you will be given homework assignments that are open-ended, requiring some interpretation, research, and diligence on your part. You are then expected to collaborate and share resources and knowledge that you have discovered with your fellow students. This feedback will be shared in class in the form of student teach-backs and presentations. This knowledge should also be shared outside of class on ```slack.com```. The instructor and all students should check ```slack.com``` at least once per day for updates. In addition to technical knowledge, independent learning and collaborative communication skills are important for SDEs in the real world.
