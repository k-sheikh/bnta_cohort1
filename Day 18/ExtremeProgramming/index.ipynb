# Extreme Programming (1 Day)

Extreme programming refers to a set of agile software development strategies that allow flexible and frequent releases of working code completed in short development cycles. It is intended to improve programmer productivity and to provide frequent development checkpoints that allow stakeholders to view progress and provide rapid feedback to the developers.

Software engineer Ken Beck first developed the Extreme Programming methodology. In October 1999, he published Extreme Programming Explained, detailing the entire method for others, and shortly thereafter the official website was launched as well.

## XP & SOLID code

XP is usually associated with object-oriented principles and SOLID.
SOLID is in fact an acronym and it stands for:

- The **S**ingle Responsibility Principle.
- The **O**pen-Closed Principle
- The **L**iskov Substitution Principle
- The **I**nterface Segregation Principle
- The **D**ependency Inversion Principle

## Main Benefits

- Greater design and feature agility where requirements are rapidly changing or not well understood at the outset
- Greater ability to demonstrate functionality to help stakeholders understand what they actually want sooner
- Can be more productive, especially for highly experienced and skilled developers

## Criticism

- Requires greater discipline and experience on the part of programmers
- Suspectable to "feature creep" and "throw-away work"
- Harder to schedule and budget for
- Greater uncertainty in ultimate functional outcome
- Design and development can become chaotic and uncoordinated
- A more organic evolution of features and functions can result in a code base with more inconsistencies and less efficiencies
- Can be difficult to make work on large projects with many developers involved
- Not suitable for projects that are well understood with well-established rigid requirements
