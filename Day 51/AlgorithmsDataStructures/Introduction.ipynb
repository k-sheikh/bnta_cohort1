# Algorithms and Data Structures: Introduction (1 Hr)

**NOTE**: This course content and the instructor provide structure and guidance for self-directed learning activities. The instructor, acting primarily as a coach, will explain and demonstrate crucial and fundamental concepts and background knowledge in lecture format using whiteboarding, web links, and code example walk-throughs. The instructor will also lead group discussion, posing and fielding relevant questions. However, students are encouraged to also research answers to their own questions wherever and whenever possible. Students are also encouraged to share their own research findings with other students in class and during lab session break-out rooms. Programming is primarily a social team activity, and effective communications skills must be developed. The ultimate goal is to develop a life-long-learning mentality that will serve the student well in a long and successful career. This represents the ideal software developer candidate that technology organizations are most interested in hiring.

## Student Responsibilities
Every student is responsible, individually, and in groups, to achieve the following skills:
- Develop disciplined, pro-active, self-driven, and self-responsible learning and communication strategies and habits
- Develop effective research skills using google.com, stackoverflow.com, quora.com, wikipedia.com, etc.
- Quickly develop prototypes and code experiments to learn and verify understanding of programming concepts and techniques
- Become comfortable and efficient in live code sharing sessions (e.g. Visual Studio Live Share)
- Build up your code portfolio on github.com and/or bitbucket.org
- Build up your challenge solutions on hackerrank.com, exercism.io, etc.
- Practice programming in multiple languages including Python, Java, C#, JavaScript, C++, and Go
- Learn how to describe algorithm complexity using Big O notation
- Collaborate and communicate algorithms using pseudo code, UML diagrams, and Whiteboarding
- Share code with other developers using VS Code, Git, Docker, Slack, Google Docs, etc.
- Instrument code to measure code performance (execution time and memory consumption)
- Implement major Sort and Search algorithms in multiple languages
- Implement major Data Structures in multiple languages
- Implement Unit Testing in multiple languages
