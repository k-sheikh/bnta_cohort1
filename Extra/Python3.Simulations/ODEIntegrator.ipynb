# ODE Integrator

* Solve Differential Equations Using ```odeint```
* See: https://youtu.be/VV3BnroVjZo

````python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
````

$\frac{dy(t)}{dt} = -ky(t)$

````python
# function that returns dy/dt
def model(y, t):
    k = 0.3
    dydt = -k * y
    return dydt

# initial condition
y0 = 5

# time points
t = np.linspace(0,20)

# solve ODE
y = odeint(model, y0, t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
````

$\frac{dy(t)}{dt} = -y(t) + 1$

$y(0) = 0$

````python
# function that returns dy/dt
def model(y,t):
    dydt = -y + 1.0
    return dydt

# initial condition
y0 = 0

# time points
t = np.linspace(0,5)

# solve ODE
y = odeint(model,y0,t)

# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
````

### Second order differential equation for angle theta of pendulum acted on by gravity and friction 

$\frac{d^2\theta}{dt^2} + \frac{b}{m}\frac{d\theta}{dt} + \frac{g}{l}sin(\theta) = 0$

See: https://services.math.duke.edu/education/ccp/materials/diffeq/pendulum/pend1.html

````python
def pend(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

b = 0.25
c = 5.0
y0 = [np.pi - 0.1, 0.0]
t = np.linspace(0, 50, 401)
sol = odeint(pend, y0, t, args=(b, c))
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
````
