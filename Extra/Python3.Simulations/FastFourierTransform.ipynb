# Fast Fourier Transform (FFT)

See: https://docs.scipy.org/doc/scipy/reference/tutorial/fft.html

The DFT (Discrete Fourier Transform) $y[k]$ of sequence $x[n]$ is defined as
$$y[k] = \sum_{n=0}^{N-1} e^{-2\pi j \frac{kn}{N}} x[n]$$
and the inverse DFT transform $x[n]$ is defined as
$$x[n] = \frac{1}{N} \sum_{k=0}^{N-1} e^{2\pi j \frac{kn}{N}} y[k]$$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

N = 600         # Number of sample points
T = 1.0 / 800.0 # sample spacing
x = np.linspace(0.0, N*T, N)
y = 2.0 * np.sin(50.0 * 2.0*np.pi * x) + 3.0 * np.sin(80.0 * 2.0*np.pi * x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
