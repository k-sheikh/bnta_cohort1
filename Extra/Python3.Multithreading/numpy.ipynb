````python
%matplotlib inline
# numpy.linalg.solve solves a linear matrix equation, or system of linear scalar equations.

# Given matrix A and vector b -> solve for vector x in: Ax = b

#   A * x   =  b
# ┌   ┐┌  ┐   ┌  ┐
# │1 2││x1│   │10│
# │3 4││x2│ = │20│
# └   ┘└  ┘   └  ┘

# Manual solution:

# 1*x1 + 2*x2 = 10  <- linear equation 1
# 3*x1 + 4*x2 = 20  <- linear equation 2

# 2*x1 + 4*x2 = 20  <- multiplied all terms in linear equation 1 by 2
# 3*x1 + 4*x2 = 20  <- linear equation 2 unchanged
# ----------------
# 2*x1 - 3*x1 = 0   <- subtract linear equation 2 from linear equation 1

# (2-3)*x1 = 0
# x1 = 0            <- solved for x

# 1*0 + 2*x2 = 10   <- substitute solution for x into linear equation 1
# x2 = 5            <- solved for y

# [x1, x2] = [0. 5]    <- solution x is [0. 5]

# Numpy solution:
from numpy.linalg import solve
A = np.array([[1,2],[3,4]])
print(A)
b = np.array([10, 20]) 
print(b)
x = solve(A,b)       # <- numpy.linalg.solve solves a linear matrix equation, or system of linear scalar equations.
print(x)             #  <- solution x is:  [0.0 5.0]

# plot linear equation 1
# 1*x1 + 2*x2 = 10   <- linear equation 1
# x2 = (10-(1*x1))/2 <- rearrange to get x2 as function of x1
x1_1 = -10
x2_1 = (10-(1*x1_1))/2
x1_2 = +10
x2_2 = (10-(1*x1_2))/2
plt.plot([x1_1, x1_2], [x2_1, x2_2], label='1*x1 + 2*x2 = 10')

# plot linear equation 2
# 3*x1 + 4*x2 = 20   <- linear equation 2
# x2 = (20-(3*x1))/4 <- rearrange to get x2 as function of x1
x1_1 = -10
x2_1 = (20-(3*x1_1))/4
x1_2 = +10
x2_2 = (20-(3*x1_2))/4
plt.plot([x1_1, x1_2], [x2_1, x2_2], label='3*x1 + 4*x2 = 20')

#plt.plot(x[0], x[1], marker="o", label='Solution: (' + str(x[0]) + ", " + str(x[1]) + ")")
plt.plot(x[0], x[1], marker="o", label=f"Solution: ({x[0]}, {x[1]})")
plt.title('Ax = b')
plt.xlabel("x1")
plt.ylabel("x2", rotation=0)

plt.legend()

plt.show()
````

````python
# https://matplotlib.org/gallery/animation/strip_chart.html#sphx-glr-gallery-animation-strip-chart-py
%matplotlib notebook

from matplotlib.lines import Line2D
import matplotlib.animation as animation

class Scope(object):
    def __init__(self, ax, maxt=2, dt=0.02):
        self.ax = ax
        self.dt = dt
        self.maxt = maxt
        self.tdata = [0]
        self.ydata = [0]
        self.line = Line2D(self.tdata, self.ydata)
        self.ax.add_line(self.line)
        self.ax.set_ylim(-.1, 1.1)
        self.ax.set_xlim(0, self.maxt)

    def update(self, y):
        lastt = self.tdata[-1]
        if lastt > self.tdata[0] + self.maxt:  # reset the arrays
            self.tdata = [self.tdata[-1]]
            self.ydata = [self.ydata[-1]]
            self.ax.set_xlim(self.tdata[0], self.tdata[0] + self.maxt)
            self.ax.figure.canvas.draw()

        t = self.tdata[-1] + self.dt
        self.tdata.append(t)
        self.ydata.append(y)
        self.line.set_data(self.tdata, self.ydata)
        return self.line,


def emitter(p=0.03):
    'return a random value with probability p, else 0'
    while True:
        v = np.random.rand(1)
        if v > p:
            yield 0.
        else:
            yield np.random.rand(1)

# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()
scope = Scope(ax)

# pass a generator in "emitter" to produce data for the update func
ani = animation.FuncAnimation(fig, scope.update, emitter, interval=10,
                              blit=True)

plt.show()
````

## The ```numpy.ndarray``` Class: N-dimensional Array 

[SyPy.org: The N-dimensional array](https://docs.scipy.org/doc/numpy/reference/arrays.ndarray.html)  

- The **ndarray** class is the NumPy array type, not the Standard Python Library one-dimensional **array.array** class.  
- An ndarray is a (usually fixed-size) multidimensional container of items of the same type and size.
- The elements in an ndarray can be accessed and modified by indexing or slicing the array.
- Different ndarrays can share the same data, so that changes made in one ndarray may be visible in another. 

Some of the ndarray object attributes:
- **ndarray.ndim** number of dimensions (rank is number of axes)
- **ndarray.shape** tuple of integers (n,m) for n rows and m columns for array size
- **ndarray.size** total number of array elements
- **ndarray.dtype** object describing type of array elements
- **ndarray.itemsize** element size in bytes
- **ndarray.data** buffer containing actual elements of the array (seldom useful)

## Broadcasting

[Array Broadcasting in numpy](http://scipy.github.io/old-wiki/pages/EricsBroadcastingDoc)  

- The term broadcasting describes how numpy treats arrays with different shapes during arithmetic operations
- NumPy operations are usually done on pairs of arrays on an element-by-element basis
- An array and a scalar value are combined on an element-by-element basis
- NumPy compares shapes element-wise and starts with the trailing dimensions, and works its way forward
- Two dimensions are compatible if they are equal, or if one of them is 1
- Broadcasting provides a convenient way of taking the outer product

````python
import numpy as np

a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
c = a * b
print(c)

a = np.array([1.0, 2.0, 3.0])
b = 2.0
c = a * b
print(c)

x = np.arange(4)
xx = x.reshape(4,1)
y = np.ones(5)
z = np.ones((3,4))

# https://docs.scipy.org/doc/numpy/user/basics.broadcasting.html
x.shape
y.shape
#x + y     # <type 'exceptions.ValueError'>: shape mismatch: objects cannot be broadcast to a single shape
xx.shape
y.shape
(xx + y).shape
xx + y
x.shape
z.shape
(x + z).shape
x + z
````

## NumPy Linear Algebra

https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.linalg.html

- **```vdot(a, b)```** Return the dot product of two vectors
- **```dot(a, b[, out])```** Dot product of two arrays
- **```inner(a, b)```** Inner product of two arrays
- **```outer(a, b[, out])```** Compute the outer product of two vectors
- **```linalg.solve(a, b)```** Solve a linear matrix equation, or system of linear scalar equations
- **```linalg.inv(a)```** Compute the (multiplicative) inverse of a matrix
- **```linalg.det(a)```** Compute the determinant of an array
- **```linalg.eig(a)```** Compute the eigenvalues and right eigenvectors of a square array
- **```linalg.eigvals(a)```** Compute the eigenvalues of a general matrix
- **```matmul(a, b[, out])```** Matrix product of two arrays

````python
# http://nbviewer.jupyter.org/github/mbakker7/exploratory_computing_with_python/blob/master/notebook6_linear_systems/py_exploratory_comp_6_sol.ipynb

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

xp = np.array([-2, 1, 4])
yp = np.array([2, -1, 4])
A = np.zeros((3, 3))
rhs = np.zeros(3)
for i in range(3):
    A[i] = xp[i] ** 2, xp[i], 1  # Store one row at a time
    rhs[i] = yp[i]
print('Array A:')
print(A)
print('rhs:',rhs)

sol = np.linalg.solve(A, rhs)
print('solution is:', sol)
print('specified values of y:', yp)
print('A @ sol:', A @ sol)

plt.plot(xp, yp, 'ro')
x = np.linspace(-3, 5, 100)
y = sol[0] * x ** 2 + sol[1] * x + sol[2]
plt.plot(x, y, 'b');
````

````python
# http://nbviewer.jupyter.org/github/mbakker7/exploratory_computing_with_python/blob/master/notebook6_linear_systems/py_exploratory_comp_6_sol.ipynb

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

tp = np.array([0, 0.25, 0.5, 0.75])
yp = np.array([ 3, 1, -3, 1])
A = np.zeros((4, 4))
rhs = np.zeros(4)
for i in range(4):
    A[i] = np.cos(1 * np.pi * tp[i]), np.cos(2 * np.pi * tp[i]), \
           np.cos(3 * np.pi * tp[i]), np.cos(4 * np.pi * tp[i])  # Store one row at a time
    rhs[i] = yp[i]
sol = np.linalg.solve(A, rhs)
print('a,b,c,d: ',sol)

t = np.linspace(0, 1, 100)
y = sol[0] * np.cos(1 * np.pi * t) + sol[1] * np.cos(2 * np.pi * t) + \
    sol[2] * np.cos(3 * np.pi * t) + sol[3] * np.cos(4 * np.pi * t)
plt.plot(t, y, 'b', label='wave')
plt.plot(tp, yp, 'ro', label='data')
plt.legend(loc='best');
````

````python
# http://nbviewer.jupyter.org/github/mbakker7/exploratory_computing_with_python/blob/master/notebook6_linear_systems/py_exploratory_comp_6_sol.ipynb

def fpoly(x, N):
    rv = np.zeros(N + 1)
    for n in range(N + 1):
        rv[n] = x ** n
    return rv

print(fpoly(2, 4))

def solvepoly(x, y):
    N = len(x) - 1
    mat = np.zeros((N+1, N+1))
    for n in range(N+1):
        mat[n] = fpoly(x[n], N)
    par = np.linalg.solve(mat, y)
    return par
    
xp = np.array([0, 1, 2, 3])
yp = np.array([3, 1, -3, 1])
a = solvepoly(xp, yp)
print(a)

def fpolyeval(x, a):
    rv = np.zeros(len(x))
    for n in range(len(a)):
        rv += a[n] * x ** n
    return rv

print(fpolyeval(xp, a))

x = np.linspace(-1,4,100)
y = fpolyeval(x, a)
plt.plot(xp, yp, 'ko', label='data')
plt.plot(x, y, label='fitted poly')
plt.legend(loc='best');
````

````python
# numpy_dot_prod.py

import numpy as np
from numpy.random import rand
from numpy.linalg import solve, inv

print("Numpy Demo")

# define 3x3 array named first_array
first_array = np.array([[1, 2, 3], [3, 4, 6.7], [5, 9.0, 5]])
print("\nfirst_array:")
print(first_array)

# calculate transpose of first_array
transpose_array = first_array.transpose()
print("\ntranspose_array:")
print(transpose_array)

# calculate inverse of first_array
inverse_array = inv(first_array)
print("\ninverse_array:")
print(inverse_array)

# solve -> first_array * x = second_array
second_array =  np.array([3, 2, 1])
x = solve(first_array, second_array)
print("\nsecond_array/first_array:")
print(x)

# create a 3x3 random matrix of values within [0,1] scaled by 20
third_array = rand(3, 3) * 20
print("\nthird_array:")
print(third_array)

# calculate dot product of first_array * third_array
dot_prod = np.dot(first_array, third_array)
print("\ndot_prod:")
print(dot_prod)
````

## Linear Matrix Transform on Vector and the ```@``` Operator

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def apply_transformation(x, color, xlim_left, xlim_right, ylim_left, ylim_right, title=None, matrix=None):
    plt.axes().set_aspect('equal')
    plt.gca().set_title(title)
    plt.gca().set_xlim(xlim_left, xlim_right) # NOTE: gca means "get the current Axes" instance
    plt.gca().set_ylim(ylim_left, ylim_right)
    if matrix is None: # default to no transformation (equivalent to unity matrix)
        y = x
    else:
        y = np.array((matrix @ x.T).T) # NOTE: matrix @ x -> is same as -> R.dot(x)
    a, b = zip(*y)
    plt.plot(a, b, color)
    plt.show()

x = np.array([[5.0, 5.0], [5.0, -5.0], [-5.0, -5.0], [-5.0, 5.0], [5.0, 5.0]]) # input array of coordinate pairs

apply_transformation(x, "red", -20, 20, -20, 20, "no transformation")             # transformation: no transformation

theta = np.radians(30)  # convert degrees to radians
c, s = np.cos(theta), np.sin(theta)
matrix = np.matrix([[c, -s], [s, c]])
apply_transformation(x, "green", -20, 20, -20, 20, "rotate 30 degrees counter-clockwise", matrix)   # transformation: rotate 30 degrees counter-clockwise

matrix = np.matrix([[1, 2], [0, 1]])
apply_transformation(x, "blue", -20, 20, -20, 20, "horizontal shear by factor of 2", matrix)    # transformation: horizontal shear by factor of 2

matrix = np.matrix([[1, 0], [2, 1]])
apply_transformation(x, "magenta", -20, 20, -20, 20, "vertical shear by factor of 2", matrix) # transformation: vertical shear by factor of 2
````
