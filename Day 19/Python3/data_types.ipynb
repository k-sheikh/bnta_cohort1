```python
# simple list comprehensions

my_numbers_list = [1, 2, 3, 4]
my_squares_list = [el ** 2 for el in my_numbers_list]

print("list of numbers: " + str(my_numbers_list))
print("list of squares: " + str(my_squares_list))

print("list of squares not equal to 9: ", end="")
print([el ** 2 for el in my_numbers_list if el ** 2 != 9])

print("another list example: ", end="")
print([2*x for x in range(100) if x >= 90 and x % 2 == 0])

#Dictionary Comprehensions

my_squares_dict = {el: el ** 2 for el in my_numbers_list}
print("dictionary of squares: " + str(my_squares_dict))

#Set Comprehensions

a_set = {1, 2, 3, 4, 5}
print({el ** 3 for el in a_set})

print({el ** 3 for el in set(range(10))})
```

```python
# More list comprehensions

text = "I am so clever that sometimes I don't understand a single word of what I am saying"

words = text.split()
word_lengths = []

print ("Word lengths of words not equal to 'clever' using for loop")
for word in words:
    if word != "clever":
        word_lengths.append(len(word))
print(word_lengths)

print ("Word lengths of words not equal to 'clever' using list comprehension")
word_lengths = [len(word) for word in words if word != "clever"]
print(word_lengths)

squares = []
for x in range(10): # using for loop
    squares.append(x**2)
print(squares)

squares = [x**2 for x in range(10)] # using comprehension instead of for loop
print(squares)
```

## Queues

```python
"""
It is possible to use a list as a queue, where the
first element added is the first element retrieved (FIFO).
However, lists are not efficient for implementing queues.
Appends and pops from the end of list are fast, but inserts
or pops from the beginning of a list is slow due to shifting.
To implement a queue, use collections.deque, designed to have
fast appends and pops from both ends. 
"""

from collections import deque
queue = deque(["Peter", "Joe", "Jenny"])
print(queue)
queue.append("Bart")
print(queue)
queue.append("Homer")
print(queue)
queue.popleft()
print(queue)
queue.popleft()
```
